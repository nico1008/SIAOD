### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
from random import randint

user_m = input()


user_n = input()
user_min_limit = input()
user_max_limit = input()
user_m, user_n, user_min_limit, user_max_limit


def outMatrix(matrix):
    for line in matrix:
        x = [str(elem) for elem in line]
        print(" ".join(x))


def genMatrix(m, n, min_lim, max_lim):
    return [[randint(int(min_lim), int(max_lim)) for j in range(int(n))] for i in range(int(m))]


myMatrix=genMatrix(user_m, user_n, user_min_limit, user_max_limit)


outMatrix(myMatrix)
```

```python
import time
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
def sort_select(line):
    for i in range (len(line)-1):
        nMin=i
        for j in range(i + 1, len(line)):
            if line[j] <line[nMin]:
                nMin=j
        line[i],line[nMin]=line[nMin],line[i]
for line in myMatrix:
    line= sort_select(line)
    
outMatrix(myMatrix)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
