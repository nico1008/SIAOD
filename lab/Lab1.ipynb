### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
from random import randint

user_m = input()


user_n = input()
user_min_limit = input()
user_max_limit = input()
user_m, user_n, user_min_limit, user_max_limit


def outMatrix(matrix):           #функция печати матрицы  
    for line in matrix:
        x = [str(elem) for elem in line]
        print(" ".join(x))


def genMatrix(m, n, min_lim, max_lim):                  #функция генерации матрицы
    return [[randint(int(min_lim), int(max_lim)) for j in range(int(n))] for i in range(int(m))]


myMatrix=genMatrix(user_m, user_n, user_min_limit, user_max_limit)


outMatrix(myMatrix)
```

Реализовать методы сортировки строк числовой матрицы в соответствии с

временем стандартной функции сортировки. Испытания проводить на сгенерированных

```python
import time
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
myMatrix1=myMatrix

start_time = time.time()
def sort_select(line):
    for i in range (len(line)-1):
        nMin=i
        for j in range(i + 1, len(line)):
            if line[j] <line[nMin]:
                nMin=j
        line[i],line[nMin]=line[nMin],line[i]
for line in myMatrix1:
    line= sort_select(line)
    
outMatrix(myMatrix1)
sort_select_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
myMatrix2=myMatrix

start_time = time.time()

def sort_insertion(line):
    
    for i in range (1,len(line)):
        
        for j in range (i-1,0,-1):                      # вставляем элемент в то место, где он минимальный 
            if (line[j]>line[j+1]):
                line[j],line[j+1]=line[j+1],line[j]



                
for line in myMatrix2:
    line= sort_insertion(line)
    
    
outMatrix(myMatrix2)

sort_insertion_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
myMatrix3=myMatrix

start_time = time.time()
def sort_bubble(line):
    for i in range (len(line)-1):
        for j in range( len(line)-i-1):                  # попарно сравниваем элементы 
            if line[j] >line[j+1]:
                
                line[j],line[j+1] = line[j+1],line[j]
                
                
                
for line in myMatrix3:
    line= sort_bubble(line)
    
outMatrix(myMatrix3)
sort_bubble_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
myMatrix4=myMatrix
start_time = time.time()
#TODO

def sort_shell(line):
    half = len(line) // 2
    while half > 0:
        for i in range(half, len(line)):
            t = line[i]
            j = i
            while j >= half and line[j - half] > t:
                line[j] = line[j -  half]
                j -=  half
 
            line[j] = t
    
        half = half // 2


for line in myMatrix4:
    line= sort_shell(line)



outMatrix(myMatrix4)

sort_shell_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
myMatrix5=myMatrix
start_time = time.time()
#TODO

def sort_quick(line):
    
    if len(line)<=1:
        return line
    elem= line[0]
    left =list(filter(lambda x:x<elem,line))                        #разбиваем строку на 3 группы 
    center = [i for i in line if i==elem]                           #продолжаем пока групп не останется
    right=list(filter (lambda y: y> elem,line))                     #рекурсия 
    
    
    return sort_quick(left) + center + sort_quick(right)
    



for line in myMatrix5:
    line= sort_quick(line)


outMatrix(myMatrix5)

sort_quick_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.

import heapq

myMatrix6 = myMatrix

start_time = time.time()

def sort_tournament(line, places):          #  places-> сколько мест в решетке 
    heap = []
    if len(line) >= places:                 #создаем кучу 
        heap = line[:places]
    else:
        heap = line[:] 
    heapq.heapify(heap)                      #heapify – список в кучу 
    
    winners = [float('-inf')]                #  2 массива с победителями ( наименьшими )и 
    losers = []                              #  и с проигравшими (наибольшими) 
    
    i = places
    while i < len(line) or len(heap) > 0:
        heap_top = heapq.heappop(heap)
        if heap_top >= winners[-1]:          #определяем в какой массив идет элемент 
            winners.append(heap_top)
        else:
            losers.append(heap_top)
        
        if i < len(line):                      # не забываем сравнивать новый элемент с элементом победителем 
            heapq.heappush(heap, line[i])      # heappush – добавить элемент в кучу   
        i += 1
    
    del winners[0]
    line.clear()               
    line.extend(losers)                       #склеиваем победителей и проигравших 
    line.extend(winners)
    
    if len(losers) > 0:
        sort_tournament(line, places)
    
for i in range(len(myMatrix6)):
    sort_tournament(myMatrix6[i], 4)
    
outMatrix(myMatrix6)

sort_tournament_time = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(sort_tournament_time))



```
